```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin'].fillna('U', inplace=True)
df['Cabin'] = df['Cabin'].apply(lambda x: x[0])
df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = sns.countplot(x = 'Survived',hue ='Sex',data = df)
```

```python
p_class = pd.get_dummies(df['Pclass'],drop_first = True)
p_class.head()
```

```python
em = pd.get_dummies(df['Embarked'],drop_first = True)
em.head()
```

```python
df_1 = pd.get_dummies(df['Sex'])
df_1.head(5)
```

```python
cabin_1 = pd.get_dummies(df['Cabin'],drop_first = True)
cabin_1.head()
```

```python
cabin_1 = cabin_1.drop(['T'],axis = 1)
```

```python
df = pd.concat([df,df_1,em,p_class,cabin_1],axis = 1)
df = df.drop(['Name','Sex','Pclass','Embarked','PassengerId','Ticket','Cabin','Fare'],axis = 1)
```

```python
df.isnull().sum()
```

```python
from sklearn.cross_validation import train_test_split
y = df['Survived']
X = df.drop('Survived',axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
logis_model = LogisticRegression()
logis_reg = logis_model.fit(X_train,y_train)
```

```python
X.head()
```

```python
logis_reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
ran_model = RandomForestClassifier(n_estimators = 100,criterion = 'entropy')
ran_for = ran_model.fit(X_train,y_train)
```

```python
ran_for.score(X_test,y_test)
```

```python
d = ran_for.predict(X_test)
p = logis_reg.predict(X_test)
from sklearn.metrics import accuracy_score,r2_score
r2_score(p,y_test)
```
